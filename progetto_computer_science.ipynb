### 1. Normalize the location field of *Users* dataset, splitting into city, region, country.

### 7. For each book in `goodbooks`, compute the list of its authors. Assuming that the number of reviews with a text (column `work_text_reviews_count`) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the *shared number of reviews with a text*.

```python
goodbooks_books.head()
```

```python
authors_name=goodbooks_books.set_index('book_id').authors.str.split(',', expand=True).stack().reset_index('book_id').rename(columns={0:'Authors'})
authors_name
```

```python
authors_name.nunique()
```

```python
df1=goodbooks_books[["book_id","work_text_reviews_count"]]
```

```python
df2=pd.merge(authors_name,df1, on="book_id")
size=df2.groupby("book_id").size().reset_index()
df2=pd.merge(df2,size, on="book_id")
df2.rename(columns={0:'size'},inplace=True)
df2
```

```python
df2["shared_number_of_reviews_with_a_text"]=df2["work_text_reviews_count"]/df2["size"]
```

```python
df2[["Authors","shared_number_of_reviews_with_a_text"]].groupby("Authors").sum()
```
